#### 1. Qual diferença a Internet faz nos estudos?

```python
soma_sem_net = 0 # soma das notas dos estudantes sem internet
soma_com_net = 0 # mesma coisa para os sem internet
quant_sem_net = 0 # quantidade acumulativa de estudantes sem internet
quant_com_net = 0 # mesma coisa para os sem internet

for index, row in df.iterrows():
    if row['Internet_Access'] == 'Yes':
        soma_com_net += row['Exam_Score']
        quant_com_net += 1
    else:
        soma_sem_net += row['Exam_Score']
        quant_sem_net += 1

media_sem = soma_sem_net/quant_sem_net
media_com = soma_com_net/quant_com_net
media_com, media_sem
```

```python
categorias = ['Sem acesso', 'Com acesso']
medias = [media_sem, media_com]

plt.figure(figsize=(4, 6))
plt.bar(categorias, medias, color=['blue', 'orange'])
plt.ylim(60, 75)
plt.xlabel('Acesso a internet')
plt.ylabel('Média')
plt.title('Média do estudantes com ou sem acesso a internet')
plt.show()
```

Como podemos ver, aparentemente a Internet não é um grande impecílio para os estudos no contexto do dataset, uma vez que a diferença de pontos no exame final é pequena

```python
print(f"a diferença média de pontos entre pessoas que tem acesso a Internet e pessoas que não tem acesso a Internet: {round(media_com - media_sem, 2)} pontos no exame final")
```

#### 2. Qual a diferença de qualidade de ensino entre instituições de ensino públicas e privadas?

```python
soma_publica = 0 # Soma das notas de estudantes instituições de ensino públicas
soma_privada = 0 # Mesma coisa para a privada
quant_alunos_publica = 0 # Quantidade de alunos de instituições de ensino públicas
quant_alunos_privada = 0 # Mesma coisa para a privada

for index, row in df.iterrows():
    if row['School_Type'] == 'Public':
        soma_publica += row['Exam_Score']
        quant_alunos_publica += 1
    else:
        soma_privada += row['Exam_Score']
        quant_alunos_privada += 1

media_publica = soma_publica / quant_alunos_publica
media_privada = soma_privada /quant_alunos_privada

media_publica, media_privada
```

```python
categorias = ['Públicas', 'Privadas']
medias = [media_publica, media_privada]

plt.figure(figsize=(4, 6))
plt.bar(categorias, medias, color=['blue', 'orange'])
plt.ylim(60, 75)
plt.xlabel('Rendimento dos alunos')
plt.ylabel('Média')
plt.title('Média dos estudantes de instituições de ensino públicas ou privadas')
plt.show()
```

No contexto desse dataset, como podemos ver, a diferença de qualidade entre instituições de ensino públicas e instituições de ensino privada não é um fator que implica no rendimento dos alunos.

```python
print(f"A diferença entre as médias dos alunos da privada e da pública é de: {round(media_privada - media_publica, 2)} pontos no exame final, uma diferença quase que irrelevante")
```

#### Rendimento dos alunos baseado na quantidade de aulas de reforço

```python
qtd_alunos_aulas_por_semana = [0, 0, 0, 0, 0, 0, 0, 0] # Quantidade de alunos que fazem x aulas de reforço por semana x[0, 8]
soma_nota_alunos = [0, 0, 0, 0, 0, 0, 0, 0] # Soma das notas dos alunos

# Conta quantos alunos e soma suas notas
for index, row in df.iterrows():
    match row['Tutoring_Sessions']:
        case 0:
            qtd_alunos_aulas_por_semana[0] += 1
            soma_nota_alunos[0] += row['Exam_Score']
        case 1:
            qtd_alunos_aulas_por_semana[1] += 1
            soma_nota_alunos[1] += row['Exam_Score']
        case 2:
            qtd_alunos_aulas_por_semana[2] += 1
            soma_nota_alunos[2] += row['Exam_Score']
        case 3:
            qtd_alunos_aulas_por_semana[3] += 1
            soma_nota_alunos[3] += row['Exam_Score']
        case 4:
            qtd_alunos_aulas_por_semana[4] += 1
            soma_nota_alunos[4] += row['Exam_Score']
        case 5:
            qtd_alunos_aulas_por_semana[5] += 1
            soma_nota_alunos[5] += row['Exam_Score']
        case 6:
            qtd_alunos_aulas_por_semana[6] += 1
            soma_nota_alunos[6] += row['Exam_Score']
        case 7:
            qtd_alunos_aulas_por_semana[7] += 1
            soma_nota_alunos[7] += row['Exam_Score']
```

```python
media_nota_alunos = [0, 0, 0, 0, 0, 0, 0, 0] # Média das notas dos alunos

# Calcula a média
for i in range(8):
    media_nota_alunos[i] = soma_nota_alunos[i] / qtd_alunos_aulas_por_semana[i]
    print(f"media dos alunos que fazem {i} aulas de reforço semanais: {round(media_nota_alunos[i], 2)} pontos no exame final")
```

```python
plt.figure(figsize=(10,6))

categorias = ["Nenhuma", "Uma aula", "Duas aulas", "Três aulas", "Quatro aulas", "Cinco aulas", "Seis aulas", "Sete aulas"]

df_test = pd.Series(qtd_alunos_aulas_por_semana, index=categorias)

df_test.plot(kind = 'barh', color = 'cyan')

plt.xlabel('Alunos')
plt.ylabel('Quantidade de aulas de reforço por semana')
plt.show()
```

```python
categorias = ["Nenhuma", "Uma aula", "Duas aulas", "Três aulas", "Quatro aulas", "Cinco aulas", "Seis aulas", "Sete aulas"]

plt.figure(figsize=(12, 6))
plt.bar(categorias, media_nota_alunos, color=['blue', 'orange'])
plt.ylim(60, 75)
plt.xlabel('Quantidade de aulas de reforço por semana')
plt.ylabel('Rendimento dos alunos')
plt.title('Aulas de reforço por semana')
plt.show()
```

A partir disso, podemos concluir que aulas de reforço extra fazem diferença consideravel na nota final do estudantes, sendo **seis aulas** o "número ótimo", talvez sete aulas sejam demais, pois é necessário que o estudante tenha tempo para estudar sozinho.

#### 4. Há diferença de disponibilidade de recursos educacionais entre estudantes de escolas públicas e privadas?

```python
# Criando uma tabela de contingência para verificar a frequência de combinações de valores entre duas colunas
tabela_contingencia = pd.crosstab(
                        df['School_Type'],
                        df['Access_to_Resources'],
                        margins = True,
                        margins_name = "Número total de alunos")
tabela_contingencia
```

```python
# Tabela com percentual comparativo
tabela_contingencia_percentual = pd.crosstab(
                                    df['School_Type'],
                                    df['Access_to_Resources'],
                                    normalize = "index",
                                    margins = True,
                                    margins_name = "Percentual de alunos"
                                    ).mul(100).round(1)
tabela_contingencia_percentual
```

```python
# Retira a linha "Percentual de alunos", pois pode gerar confusão no gráfico
tabela_sem_margem = tabela_contingencia_percentual.drop("Percentual de alunos")

# Cria o gráfico de barras empilhadas
fig, ax = plt.subplots(figsize=(10, 6))

cores = ['#77ff6b', '#ff6161', '#f7ff61']

for i, access_type in enumerate(tabela_sem_margem.columns):
    bottom_values = tabela_sem_margem.iloc[:, :i].sum(axis=1) if i > 0 else 0
    bars = ax.bar(
        tabela_sem_margem.index,
        tabela_sem_margem[access_type],
        label=access_type,
        bottom=bottom_values,
        color=cores[i]
    )

    # Adiciona os valores nas barras
    for bar, value in zip(bars, tabela_sem_margem[access_type]):
        height = bar.get_height() + bar.get_y()  # Posição final da barra (cima da pilha)
        ax.text(
            bar.get_x() + bar.get_width() / 2,  # Centraliza o texto na barra
            height,  # Coloca o texto logo acima da barra
            f"{value:.1f}%",  # Formato do texto (percentual com 1 decimal)
            ha="center", va="bottom", fontsize=10
        )

ax.set_title("Distribuição percentual por tipo de escola e acesso a recursos", fontsize=12)
ax.set_ylabel("Percentual (%)")
ax.set_xlabel("Tipo de escola")
ax.legend(title="Acesso a recursos")
plt.tight_layout()
plt.show()
```

Apesar da porcentagem de alunos com disponibilidade de recursos alta ser 3,4% maior nas escolas privadas, pode-se dizer que o tipo de escola (pública ou privada) não interfere muito no acesso a recursos nesse dataset. É notório que em ambas as intituições de ensino a disponibilidade de materiais de estudo é considerado médio.

#### 7. Analisando horas de estudo e seu impacto

##### 7.1. Distribuição das horas de estudo por nota no exame

```python
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import gaussian_kde

x = df['Exam_Score']
y = df['Hours_Studied']


xy = np.vstack([x, y])
kde = gaussian_kde(xy)
x_grid, y_grid = np.meshgrid(np.linspace(60, 75, 30),
                             np.linspace(10, 30, 30))
z_grid = kde(np.vstack([x_grid.ravel(), y_grid.ravel()])).reshape(x_grid.shape)

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x_grid, y_grid, z_grid, cmap='viridis', alpha=1)
ax.set_xlabel('Nota no exame')
ax.set_ylabel('Horas estudadas')
ax.zaxis.set_visible(False)

# Esconder eixo Z
ax.zaxis.set_ticks([])
ax.zaxis.set_ticklabels([])
ax.zaxis.line.set_color((0, 0, 0, 0))
plt.title('Distribuição 3d da frequência de horas estudadas e das notas no exame')


# Mudar angulo para vista de cima
elevation = 90
azimuth = 90
ax.view_init(elev=elevation, azim=azimuth)


plt.show()
```

podemos ver no gráfico a cima que há uma pequena melhora na nota em função da quantidade de estudo vista na frequência de pessoas que estudam mais de 22.5 horas semanais


##### 7.2. Distribuição das horas de sono por horas de estudo (até que ponto vale a pena trocar o sono por estudos?)

```python
x = df['Sleep_Hours']
y = df['Hours_Studied']


xy = np.vstack([x, y])
kde = gaussian_kde(xy)
x_grid, y_grid = np.meshgrid(np.linspace(4, 10, 6),
                             np.linspace(10, 30, 20))
z_grid = kde(np.vstack([x_grid.ravel(), y_grid.ravel()])).reshape(x_grid.shape)

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x_grid, y_grid, z_grid, cmap='viridis', alpha=1)
ax.set_xlabel('Horas de sono médio por noite')
ax.set_ylabel('Horas estudadas')
ax.zaxis.set_visible(False)

# Esconder eixo Z
ax.zaxis.set_ticks([])
ax.zaxis.set_ticklabels([])
ax.zaxis.line.set_color((0, 0, 0, 0))
plt.title('Distribuição 3d da frequência de horas estudadas e das notas no exame')


# Mudar angulo para vista de cima
elevation = 45
azimuth = 90
ax.view_init(elev=elevation, azim=azimuth)

plt.show()
```

pelo que pode-se observar, parece haver um pequeno *tradeof* entre a quantidade de horas de estudo e a quantidade de sono, podemos ver que a distribuição conjunta é um pouco mais deslocada e elevada no sentido de **mais horas de sono e menos horas de estudo**, sendo assim, podemos observar uma frequência em estudantes que dormem mais em estudar menos (já que sobra menos tempo aos mesmos para realizar tal). Além dessa análise podemos observar no pico do gráfico que a maioria dos estudantes **estuda entre 17 e 22 horas semanais** e **dorme 7 horas por dia**

##### 7.3. analisando diferentes distribuições de notas em função do sono

#### 8. Matriz Correlacional
